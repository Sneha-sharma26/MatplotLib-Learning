```python
# To add basic texts to the plot like title, xlabel, ylabel, text at an arbitrary position, annotate & legend

import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("Python",fontsize= 15)
plt.xlabel("Days")
plt.ylabel("Growth")

plt.text(2,5,"java",style= "italic", bbox= {"facecolor":"r"})

plt.annotate("python",xy = (3,5), xytext=(3.5,4), arrowprops= dict(facecolor= "r",shrink= 10))

plt.legend(["C++"],loc = 1, edgecolor = "y", framealpha = 0.8, shadow= True)        # framealpha is used to lighen up the shade of the edgecolor

plt.show()
```
