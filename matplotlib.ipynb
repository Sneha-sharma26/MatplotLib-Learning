# Matplotlib Introduction

### Bar Plot

```python
import matplotlib.pyplot as plt
x = ["Python","R","Java","C"]
y = [85,55,82,70]

# Adding labels to the bar graph
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Languages(in percentage)",fontsize = 20)   

plt.bar(x,y)
plt.show()
```

```python
# changing appearance of the bars

import matplotlib.pyplot as plt
x = ["Python","R","Java","C"]
y = [85,55,82,70]

plt.bar(x,y, width=0.4, color='y')        
# width=> changes the width of the bars(0-1), 
# color=> alters the color of the bars from blue(default) to yellow
plt.show()
```

```python
# Passing a list of colors of the bars

import matplotlib.pyplot as plt
x = ["Python","R","Java","C"]
y = [85,55,82,70]

c = ['r','b','g','y']

# changing appearance of the bar graph
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Languages(in percentage)")

plt.bar(x,y, width=0.4, color=c)        
plt.show()
```

```python
# Aligning the labels  (default = center)

import matplotlib.pyplot as plt
x = ["Python","R","Java","C"]
y = [85,55,82,70]

# changing appearance of the bar graph
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Languages(in percentage)")

plt.bar(x,y, width=0.4, align='edge')        
plt.show()
```

```python
# Boundaries of the bars

import matplotlib.pyplot as plt
x = ["Python","R","Java","C"]
y = [85,55,82,70]

plt.bar(x,y, width=0.4, edgecolor='red', linewidth=4, linestyle=':', alpha=0.7)    
# linewidth => defines the width of the borders of the bars     
# linestyle => defines the border line type (solid, dotted, etc.)
# alpha     => defines the dullness/brightness of the bars (0-1)
plt.show()
```

```python
# To put label on the graph using legend() func

import matplotlib.pyplot as plt
x = ["Python","R","Java","C"]
y = [85,55,82,70]

plt.bar(x,y, width=0.4, label="y-axis = 10%,\nx-axis") 
plt.legend()   
plt.show()
```

```python
# Plotting 2 bar graph at a time

import matplotlib.pyplot as plt
import numpy as np

x = ["Python","R","Java","C"]
y = [85,55,82,70]
z = [20,30,40,10]

width = 0.2      # fix the width of the bars
p=np.arange(len(x))    # p=[0,1,2,3]
p1 = [j+width for j in p]      # p1=[0.2, 1.2, 2.2, 3.2]

plt.bar(p,y, width, label="y")
plt.bar(p1,z, width, label="z")

# without below code, the x's labels will be according to p
plt.xticks(p+width/2, x)    # xticks => Set the current labels of the x-axis.

plt.legend()   
plt.show()
```

```python
# To rotate the x's labels & moving the location of the graph's label(using legend function)

import matplotlib.pyplot as plt
import numpy as np

x = ["Python","R","Java","C"]
y = [85,55,82,70]
z = [20,30,40,10]

width = 0.2      # fix the width of the bars
p=np.arange(len(x))    # p=[0,1,2,3]
p1 = [j+width for j in p]      # p1=[0.2, 1.2, 2.2, 3.2]

plt.bar(p,y, width, label="y")
plt.bar(p1,z, width, label="z")

# without below code, the x's labels will be according to p
plt.xticks(p+width/2, x, rotation=90)    # xticks => Set the current labels of the x-axis.

plt.legend(loc='center')   
plt.show()
```

```python
# Plotting bar graph/s horizontally using 'barh' function

import matplotlib.pyplot as plt
import numpy as np

x = ["Python","R","Java","C"]
y = [85,55,82,70]
z = [20,30,40,10]

width = 0.2      # fix the width of the bars
p=np.arange(len(x))    # p=[0,1,2,3]
p1 = [j+width for j in p]      # p1=[0.2, 1.2, 2.2, 3.2]

plt.barh(p,y, width, label="y")
plt.barh(p1,z, width, label="z")

# without below code, the x's labels will be according to p
plt.xticks(p+width/2, x)    # xticks => Set the current labels of the x-axis.

plt.legend()   
plt.show()
```

# Scatter Plot 

```python
# Basic Code

import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [3,4,2,5]

plt.scatter(x,y)
plt.show()
```

```python
# Adding labels to the bar graph & ohter changes to scatter plot

import matplotlib.pyplot as plt
day = [1,2,3,4]
no = [3,4,2,5]

plt.title("Scatter Plot", fontsize = 16)   
plt.xlabel("Days")
plt.ylabel("Number")

plt.scatter(day, no, c='r', s = 100, alpha = 0.5, marker = '*' )       # color/c => to change the scatter dots/markers color  
# Instead of any specific color for every dot, a list of 4 colours can be passed as in bar graph
# s => this will alter the size of the scatter dots. Similar to color, a list of size can be passed for each dot
# alpha => defines the dullness/brightness of the scatter dots (0-1)
# marker => this will change the marker from circle/dots to defined marker(star). By default, circle
plt.show()
```

```python
# More decorations for scatter plots' markers

import matplotlib.pyplot as plt
day = [1,2,3,4]
no = [3,4,2,5]

plt.title("Scatter Plot", fontsize = 16)   
plt.xlabel("Days")
plt.ylabel("Number")

plt.scatter(day, no, c='r', s = 120, edgecolor='g', linewidth = 2, linestyle=':')  
# edgecolor => changes the edgecolor of the markers to the defined color 
# linewidth => defines the width of the border of markers 
# linestyle => defines the type of line i.e., solid, dot (by default, solid)
plt.show()
```

```python
# Setting marker colour using 'cmap'

# 'cmap' (ColorsMap) can be used to define the markers colors
# plt.colorbar() is used to get the color range bar with scatter plot

import matplotlib.pyplot as plt
day = [1,2,3,4]
no = [3,4,2,5]

plt.title("Scatter Plot", fontsize = 16)   
plt.xlabel("Days")
plt.ylabel("Number")

colors = [10, 29, 56, 73]

plt.scatter(day, no, c = colors, s = 120, cmap='viridis')
t = plt.colorbar()
t.set_label("ColorBar", fontsize = 18)  
plt.show()
```

```python
# Plotting 2 Scatter Plots

import matplotlib.pyplot as plt
day = [1,2,3,4]
no = [3,4,2,5]
no2 = [7,3,5,1]

plt.title("Scatter Plot", fontsize = 16)   
plt.xlabel("Days")
plt.ylabel("Number")

colors = [10, 29, 56, 73]
plt.scatter(day, no, c = colors, s = 120, cmap='viridis')
plt.scatter(day, no2, c = 'r', s = 80)

t = plt.colorbar()
t.set_label("ColorBar", fontsize = 18)  
plt.show()
```

## Pie Chart

```python
# basic pie chart 
# and to change attributes of the labels using "textprops" parameter (take argument as dictionary)
# & to place the values of 'x' counterclockwise in the pie chart using "counterclock" parameter
# to rotate the labels

import matplotlib.pyplot as plt

x = [10,20,30,40]
# Labels
y = ["C", "C++", "Java", "Python"]  

# plt.pie(x, labels = y, textprops = {"fontsize":16}, counterclock=True)
plt.pie(x, labels = y, textprops = {"fontsize":16}, counterclock=False, rotatelabels = True)

plt.show()
```

```python
# to show some part of the pie chart differently using "explode" parameter
# and use of "shadow" parameter
# to increase the size of the pie chart using "radius" parameter
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["C", "C++", "Java", "Python"]  

ex = [0.4, 0.0, 0.0, 0.0]
plt.pie(x, labels = y, explode = ex, shadow = True, radius = 1.3)

plt.show()
```

```python
# to change the colors of pies in the chart
# and to change the distance of the labels from the pie chart  (ideal 1.1 or 1.2)
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["C", "C++", "Java", "Python"]  

c = ['y', 'r', 'g', 'b']
# plt.pie(x, labels = y, colors = c, labeldistance = 1.6)   # this will move the labels away from the pie chart
plt.pie(x, labels = y, colors = c, labeldistance = 0.4)     # this will move the labels into the pie chart itself

plt.show()
```

```python
# To show percentage of the pies in the chart using "autopct" parameter
# and to change the starting angle from 0 to something else using "startangle" parameter
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["C", "C++", "Java", "Python"]  

plt.pie(x, labels = y, autopct = "%0.2f%%", startangle = 90)       # o.2f represents the floating no. upto 2 decimal places
plt.show()
```

```python
# to change chart's edge properties using "wedgeprops"
# edge properties will be shown when 'shodow=True'

import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["C", "C++", "Java", "Python"]  

ex = [0.4, 0.0, 0.0, 0.0]
plt.pie(x, labels = y, explode = ex, shadow=True, wedgeprops={"linewidth":8, "edgecolor":"m"})

plt.show()
```

```python
# Adding label to the pie chart using legend()
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["C", "C++", "Java", "Python"]  

plt.pie(x, labels = y)  
plt.title("Pie Chart")
plt.legend(loc = 1)   # loc attribute is used to define the position/location of the label
plt.show()
```

### Donut Pie Chart/Multiple Pie Chart

### Ring Pie Chart

```python
# For Ring Pie Chart -> Pie chart + Dot Pie Chart
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["C", "C++", "Java", "Python"]  

plt.pie(x, labels = y, radius = 1.5)   
plt.pie([1], colors = "w")  

plt.show()
```

```python
# Another method is using a circle inside a pie chart
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["C", "C++", "Java", "Python"]  

plt.pie(x, labels = y, radius = 1.5)   
cr = plt.Circle(xy=(0,0), radius = 1, facecolor="w")
plt.gca().add_artist(cr)

plt.show()
```

## Stem Plot

```python
# Basic Stem plot and changing line and marker format using "linefmt" & "markerfmt"
# to make the bottom line start from some other number instead of 0 (default)
# to change the bottom/base line colour
# adding label using legend()

import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt = "-.", markerfmt = "r+", bottom = 3, basefmt = 'g', label = "python")     # r+ => red & + symbol
plt.legend()
plt.show()
```

```python
# Changing base lines' colour using 'use_line_collection' parameter
# Changing orientation of the stem plot(horizontal & vertical(default))

import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, orientation='horizontal')    
plt.show()
```

## Box/Whisker Plot
Whiskey Plot are the upper and lower line which shows maximum & minimum.

```python
# Basic Box/Whisker Plot & changing the shape of the box using "notch" (By default -> box)

import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]

plt.boxplot(x, notch = True)
# Yellow line is the median line
plt.show()
```

```python
# To make the box horizontal using "vert" property
# To change the width of the box plot using "widths" parameter
# To change the label of the box plot using "tick_labels"
# To fill color in the box using "patch_artist" (by default -> False)

import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]

plt.boxplot(x, vert = False, widths = 0.4, tick_labels = ["python"], patch_artist = True)
plt.show()
```

```python
# To show the mean value using "showmeans" parameter (represented by green triangle)
# To show a outliers(oval shaped) example by value 120

import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, showmeans = True)

plt.show()
```

```python
# To join the outlier to the plot using "whis"

import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, showmeans = True, whis = 2)

plt.show()
```

```python
# To change the symbol and colour of outlier using "sym"

import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,sym = "g+" )
# 'g' for green colour and '+' represents symbol type
# if we keep sym=" " => this will hide the outlier

plt.show()
```

```python
# to change box, caps(min and max line) & whiskers properties using "boxprops", "capprops" & "whishkerprops" (takes i/p as dictionary)
# to change the outlier colour using "flierprops" 

import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x, boxprops= dict(color="y"), capprops= dict(color = "b"), whiskerprops= dict(color = "r"), flierprops= dict(markeredgecolor= "g"))
plt.show()
```

```python
# To plot Multiple box plots together

import matplotlib.pyplot as plt
x1 = [10,20,30,40,50,60,70]
x2 = [10,20,30,40,50,60,90]

y = [x1, x2]

plt.boxplot(y, tick_labels = ["python","java"], boxprops= dict(color="y"), capprops= dict(color = "b"), whiskerprops= dict(color = "r"))
plt.show()
```

## Area or Stack Plot

```python
# Basic Stack plot
# To add Labels to the plot using "labels" parameter & legend function
# To change the area colors by passing a list to "colors" parameter

import matplotlib.pyplot as plt

x= [1,2,3,4,5]
area1 = [1,4,2,3,2]
area2 = [2,4,1,3,4]
area3 = [2,4,3,5,2]

l = ["Area1", "Area2", "Area3"] # as we've multiple labels, so we'll pass labels value as list

# plt.stackplot(x,area)  # for a single stack plot
plt.stackplot(x, area1, area2, area3, labels = l, colors = ["r", "m","y"])
plt.legend(loc = 2)
plt.show()
```

```python
# To change the baseline value to zero, sym, wiggle & weighted_wiggle' (by default -> 0)
# to add title to the plot using "title" function
# to add x & y labels using 'xlabel' & 'ylabel' function

import matplotlib.pyplot as plt

x= [1,2,3,4,5]
area1 = [1,4,2,3,2]
area2 = [2,4,1,3,4]

# plt.stackplot(x, area1, area2, baseline ="zero")
plt.stackplot(x, area1, area2, baseline ="sym")
# plt.stackplot(x, area1, area2, baseline ="wiggle")
# plt.stackplot(x, area1, area2, baseline ="weighted_wiggle")

plt.title("Stack Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

## Step Plot

```python
# Basic Step Plot

import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.step(x,y)

plt.title("Step Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.grid()
plt.show()
```

```python
# to add marker (to show the points) using "marker" & change it's size using "ms"
# to change the color of the plot using "color" parameter
# to only change the marker face color using "mfc"

import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,4,3,2,5]

plt.step(x,y, marker="o",ms= 7, color = 'm', mfc = "b", label = "python")
plt.grid()
plt.legend()
plt.show()
```

## Fill_Between Plot

```python
# Basic Plot & gave complete x & y axis for filling the plot
# lighten the filling shade of the plot using "alpha" parameter
# using 'where' parameter

import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

plt.plot(x,y)

plt.fill_between(x,y, alpha = 0.5)       # giving complete x & y axis without detemining values using 'where'
# plt.fill_between(x,y, where= (x >= 2)&(x <= 4))      # determining x & y values

plt.title("Fill Between Plot")
plt.show()
```

```python
# Determining x & y axis values without using "where"
# changing color of the plot filling

import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)

# plt.fill_between(x= [3,4],y1= 2, y2 = 4, color = 'g')    # Since y1 and y2 are scalars, they act like: y1 = [2, 2] ,y2 = [4, 4]
plt.fill_between(x= [3,4],y1= [2,3], y2 = [2,4])    
# At x=3: filling between y1=2 and y2=2 → zero height (nothing filled).
# but At x=4: filling between y1=3 and y2=4 → fills between y=3 and y=4.

plt.title("Fill Between Plot")
plt.show()
```
