```python
# To Plot a basic linear graph 

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,2,4,1,5]

plt.plot(x,y)
plt.show()
```

```python
# to change the labels on x & y axis as per our requirements using "xticks" & "yticks" parameter
# add labels on any of the axis other then values

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,2,4,1,5]

plt.plot(x,y)

plt.xticks(x)
plt.yticks(x, labels = ["Py","Java","C","C++","R"])

plt.show()
```

```python
# Limiting the range of the labels of the axis using "xlim()" & "ylim()" functions

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,2,4,1,5]

plt.plot(x,y)

plt.xlim(0,10)
plt.ylim(0,6)

plt.show()
```

```python
# Limiting the range of the labels of the axis using "axis()" function

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,2,4,1,5]

plt.plot(x,y)

plt.axis([0,10,0,6])

plt.show()
```
