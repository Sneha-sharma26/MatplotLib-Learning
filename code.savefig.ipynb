## Savefig in Matplotlib

To save the graphs/plots externally

```python
# Saving a basic plot

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,3,1,4,5]

plt.plot(x,y)
plt.savefig("Line")
plt.show()
```

```python
# Saving using "dpi" parameter (dots per inch)

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,3,1,4,5]

plt.plot(x,y)
plt.savefig("Line1", dpi = 2000)
plt.show()
```

```python
# To change the face color of the saved image(basically to change padding color)(be default-> white)

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,3,1,4,5]

plt.plot(x,y)
plt.savefig("Line2", facecolor= "g")
plt.show()
```

```python
# to change the format of the saved file (pdf,jpg, etc.) (by default-> png)

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,3,1,4,5]

plt.plot(x,y)
plt.savefig("Line3.pdf", facecolor= "g")
plt.show()
```

```python
# To change the background color of the plot as per padding color(by default -> False (white))

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,3,1,4,5]

plt.plot(x,y)
plt.savefig("Line4", facecolor = 'g', transparent = True)
plt.show()
```

```python
# to change the ratio of the box of the plot using "bbox_inches" parameter 

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,3,1,4,5]

plt.plot(x,y)
plt.savefig("Line5", bbox_inches= "tight")    # removes the extra padding region
plt.show()
```
